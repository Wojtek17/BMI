```python
import pandas as pd 
import numpy as np 
import json
import datetime
```

```python
def get_data(comp):
    append_data=[]
    get_range=range(1,7)
    for i in get_range:
        df=pd.read_html(f"https://www.biznesradar.pl/notowania-historyczne/{comp},{i}")[0]
        df=df[["Data", "Zamknięcie"]]
        append_data.append(df)
    append_data=pd.concat(append_data)
    return append_data
```

```python
comp_list=["WIG","ALE", "ASSECO-POLAND", "DNP", "JSW-JASTRZEBSKA-SPOLKA-WEGLOWA", "CD-PROJEKT", "KGHM", "KRUK", "PCO", "PGE", "PZU", "PKO", "SPL", "MBANK", "ALIOR-BANK","KETY","LPP", "ORLEN", "PEKAO","ORANGE", "CYFROWY-POLSAT"]
```

```python
wig_dict={}
for comp in comp_list:
    try:
        df=get_data(comp)
        wig_dict[comp]=df
    except Exception as e:
        print(comp, e)
```

```python
json_dict = {}
for key, value in wig_dict.items():
    json_dict[key] = value.to_dict(orient='records')
```

```python
with open("/Users/wojtekmarszalek/Desktop/wig.json", 'r') as file:
    json_data = file.read()

json_dict = json.loads(json_data)

wig_dict = {}
for key, value in json_dict.items():
    wig_dict[key] = pd.DataFrame.from_records(value)
```

```python
for key, df in wig_dict.items():
    df.rename({"Zamknięcie":key}, axis=1, inplace=True)
    
df_combined=pd.concat(wig_dict.values(), axis=1)

selected_columns = [df_combined.columns[i] for i in range(1, len(df_combined.columns), 2)]
df_selected = df_combined[selected_columns]
df_selected["Data"]=df_combined.iloc[:,0]
df_selected=df_selected[['Data','WIG','ALE', 'ASSECO-POLAND', 'DNP', "JSW-JASTRZEBSKA-SPOLKA-WEGLOWA", 'CD-PROJEKT', 'KGHM', 'KRUK',
       'PCO', 'PGE', 'PZU', 'PKO', 'SPL', 'MBANK', 'ALIOR-BANK', 'KETY',
       'LPP', 'ORLEN', 'PEKAO', 'ORANGE', 'CYFROWY-POLSAT']]
df_selected["Data"]=df_selected["Data"].str.replace(".","-")
df_selected["Data"]=pd.to_datetime(df_selected["Data"], format='%d-%m-%Y')
df_selected.sort_values("Data", inplace=True)
df_selected=df_selected[31:]
```
