```python
import requests as req
import pandas as pd
from bs4 import BeautifulSoup
import pdfkit
from fpdf import FPDF
```

```python
url="https://www.realmadryt.pl/"
```

```python
response=req.get(url)
html_content=response.text
soup=BeautifulSoup(html_content,'html.parser')
main_article=soup.find("div", {"class":"c-news-main-banner c-news-main-banner box-full_width-grid"})
lower_articles=soup.find_all("li",{"class":"column-width_3 rwd-break_portrait_iphone6-cut_half rwd-break_tablet-cut_half rmpl-animation opacity"})
lowest_articles=soup.find_all("div",{"class":"rmpl-flex-item"})
```

```python
main_links_list=[]
for element in main_article:
    try:
        main_a_elements=element.find_all("a",{"class":"c-news-mainPage-banner-photo box-full_width-image-relative rmpl-animation opacity"})
        for a in main_a_elements:
            main_article_link=a.get("href")
            main_links_list.append(main_article_link)
        other_a_elements=element.find_all("a",{"class":"shift-margin_1-horizontal"})
        for a in other_a_elements:
            other_main_link=a.get("href")
            main_links_list.append(other_main_link)
    except AttributeError:
        pass
main_links_list
```

```python
lower_links_list=[]
for element in lower_articles:
    link=element.find("a").get("href")
    lower_links_list.append(link)
lower_links_list
```

```python
lowest_links_list=[]
for element in lowest_articles:
    try:
        link=element.find("a").get("href")
        lowest_links_list.append(link)
    except AttributeError:
        pass
lowest_links_list
```

```python
all_articles_links=main_links_list+lower_links_list+lowest_links_list
```

```python
all_articles_links
```

```python
print("\033[1m" +"List of articles"+ "\033[0m")
for link in all_articles_links:
    print(link[link.find("-")+1:].replace("-"," "))
print("\033[1m" +"Articles"+ "\033[0m")
for link in all_articles_links:
    response=req.get(url+link)
    html_content=response.text
    soup=BeautifulSoup(html_content,'html.parser')
    h1=soup.find("h1").text
    print("\033[1m" + h1 + "\033[0m")
    try:
        div_element = soup.find('div', class_='article-content-regular acr-infinite-ad')
        p_elements=div_element.find_all("p")
        separator = soup.find('p', style='text-align:center')
    except AttributeError:
        pass
    for p in p_elements:
        if p.find_next_sibling() == separator:
            break
        print(p.text)
```

```python
with open('exported_articles.txt', 'w', encoding='utf-8',errors='ignore') as file:
    for link in all_articles_links:
        response=req.get(url+link)
        html_content=response.text
        soup=BeautifulSoup(html_content,'html.parser')
        h1=soup.find("h1").text
        file.write(h1 +"\n")
        try:
            div_element = soup.find('div', class_='article-content-regular acr-infinite-ad')
            p_elements = div_element.find_all("p")
            separator = soup.find('p', style='text-align:center')
        except AttributeErorr:
            pass
        for p in p_elements:
            if p.find_next_sibling() == separator:
                break
            file.write(p.text+"\n")
```
